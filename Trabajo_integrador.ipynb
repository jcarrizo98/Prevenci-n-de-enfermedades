El dataset est√° compuesto por 14 campos y contiene registros de 270 pacientes. A continuaci√≥n se describe cada uno de ellos:

1. Age: indica la edad del paciente. No contiene ning√∫n campo nulo.
2. Sex: Indica el g√©nero del paciente. Asumimos valor 1 = masculino, valor 0 = femenino. No contiene ning√∫n campo nulo.
3. Chest pain type: Tipo de dolor de pecho con valores del 1 al 4. Tipo de dato entero. No contiene campos nulos. Clasificaci√≥n:
1 = Typical angina: Dolor tor√°cico t√≠pico relacionado con isquemia card√≠aca (falta de flujo sangu√≠neo al coraz√≥n). Alto riesgo de enfermedad coronaria.
2 = Atypical angina: Dolor tor√°cico que no cumple completamente con las caracter√≠sticas cl√°sicas de angina. Menor riesgo de enfermedad coronaria.
3 = Non-anginal pain: Dolor en el pecho que no est√° relacionado con el coraz√≥n. Bajo riesgo de enfermedad coronaria.
4 = Asymptomatic: El paciente no experimenta dolor tor√°cico.
4. BP: Valores de presion sist√≥lica (etapa de compresi√≥n del coraz√≥n) del paciente, en unidades de mmHG. No contiene datos nulos.
5. Cholesterol: Indica el nivel de colesterol total de los pacientes, medido en mg/DL. No contiene ning√∫n campo nulo.
6. FBS over 120: Indica si el nivel de glucosa en ayunas es mayor a 120 mg/DL, con un 1 = "s√≠", y 0 = "no". No contiene campos nulos.
7. EKG results: campo num√©rico con rango de valores entre 0-2. Indica si los resultados del electrocardiograma fueron normales (0), hubo alteraciones menores (1) o mayores (2).
8. Max HR: Campo num√©rico que indica las m√°ximas pulsaciones por minuto alcanzadas. No contiene datos nulos
9. Exercise angina: Campo num√©rico que indica si el paciente experiment√≥ (valor = 1) o no experiment√≥ molestias/dolores durante el estudio (valor = 0).
10. *ST depression: Indica el valor de depresi√≥n del segmento ST con un rango de 0 (sin alteraci√≥n) a valores de 6,2 como m√°ximo en el caso particular de esta muestra de pacientes. No contiene datos nulos.
11. Slope of ST: Indica la pendiente del segmento ST, habiendo tres valores posibles, ascendente (valor = 1), plana (valor = 2, condici√≥n normal), descendente (valor = 3). No contiene datos nulos.
12. Numbers of vessels fluro: Campo num√©rico que indica el n√∫mero de vasos sangu√≠neos coronarios que presentan obstrucciones visibles. No presenta datos nulos.
13. Thallium: Indica los resultados de una prueba de perfusi√≥n mioc√°rdica con talio, que es una t√©cnica de diagn√≥stico usada para evaluar la circulaci√≥n sangu√≠nea al coraz√≥n. No presenta datos nulos.
14. Heart Disease: Indica presencia o ausencia de enfermedad card√≠aca. Tipo de dato: object, no num√©rico.

*Aclaraci√≥n del significado del segmento ST: Se eval√∫a en los electrocardiogramas e indica si el coraz√≥n es capaz de relajarse luego de la contracci√≥n evitando la tetanizaci√≥n o patolog√≠as como la isquemia. En resumen si dicho segmento se encuentra en forma de meseta, no hay riesgo. En caso de existir cierta pendiente puede indicar la presencia de una enfermedad card√≠aca.

# Codificaci√≥n de variables categ√≥ricas


```python
#Cambiar el valor num√©rico de la columna Sex y Slope of ST
data['Sex'] = data['Sex'].replace({0: 'Femenino', 1: 'Masculino'})
data['Slope of ST'] = data['Slope of ST'].replace({1: 'Ascendente', 2: 'Normal', 3: 'Descendente'})
data[['Sex','Slope of ST']]
```

# Feature selection

El m√©todo a aplicar ser√° de aprendizaje supervisado, por lo que dividiremos las variables predictoras de la variable a predecir:

```python
x = data.drop(columns=['Heart Disease'])
y = data['Heart Disease']
print(x)
```

```python
#Elimino features constantes con varianza cero
from sklearn.feature_selection import VarianceThreshold
constant_filter = VarianceThreshold(threshold=0)
constant_filter.fit(x)
```

```python
constant_columns = [column for column in x.columns
                    if column not in x.columns[constant_filter.get_support()]]

print(len(constant_columns))
```

No tenemos valores de variables predictoras con varianza cero.

## Correlaci√≥n

**Pearson**

Evaluamos los coeficientes de correlaci√≥n entre las variables.

```python
#Correlaci√≥n de Pearson
correlacion_pearson = data.corr(method='pearson')
plt.figure(figsize=(18,18))
sns.heatmap(correlacion_pearson , annot=True, cmap="viridis",annot_kws={"size":8})
plt.show()
```

```python
# Selecciona las caracter√≠sticas con coeficientes de correlaci√≥n mayores a 0.1 con respecto a 'target'
umbral_correlacion = 0.1
correlacion_target = correlacion_pearson['Heart Disease']
caracteristicas_seleccionadas = correlacion_target[correlacion_target.abs() > umbral_correlacion].index

print("Caracter√≠sticas seleccionadas con correlaci√≥n mayor a 0.1 con respecto a 'target':")
print(caracteristicas_seleccionadas)
```

**Spearman**

```python
# Calcula la correlaci√≥n de Spearman
correlacion_spearman = data.corr(method='spearman')
# Visualiza la matriz de correlaci√≥n de Spearman
plt.figure(figsize=(10, 8))
sns.heatmap(correlacion_spearman, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlaci√≥n - Spearman')
plt.show()
```

```python
umbral_correlacion = 0.1
correlacion_target = correlacion_spearman['Heart Disease']
caracteristicas_seleccionadas = correlacion_target[correlacion_target.abs() > umbral_correlacion].index

print("Caracter√≠sticas seleccionadas con correlaci√≥n mayor a 0.1 con respecto a 'target':")
print(caracteristicas_seleccionadas)
```

**Information gain**

```python
from sklearn.feature_selection import mutual_info_regression
import numpy as np
information_gains = mutual_info_regression(x, y)
```

```python
# Obtiene los √≠ndices ordenados por Information Gain en orden descendente
indices_ordenados = np.argsort(information_gains)[::-1]

# Ordena los Information Gain en el mismo orden
information_gains_ordenados = information_gains[indices_ordenados]

# Ordena el nombre de las caracter√≠sticas en el mismo orden
nombres_caracteristicas_ordenados = x.columns[indices_ordenados]

# Crea un gr√°fico de barras para visualizar los Information Gain
plt.figure(figsize=(20, 6))
plt.bar(range(len(information_gains_ordenados)), information_gains_ordenados, color='skyblue')
plt.xlabel('Caracter√≠sticas')
plt.ylabel('Information Gain')
plt.title('Information Gain de cada Caracter√≠stica (ordenado)')
plt.xticks(np.arange(len(information_gains_ordenados)), nombres_caracteristicas_ordenados, rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Valores de Information Gain para seleccionar caracter√≠sticas con un valor por encima 0.1
umbral = 0.1
caracteristicas_seleccionadas_final = x.columns[information_gains > umbral]
print(caracteristicas_seleccionadas_final)
```

Debido a que el m√©todo de Information Gain es el m√°s adecuado para variables target categ√≥ricas como lo es el caso de estudio, se utilizan las caracter√≠sticas seleccionadas de este m√©todo.

```python
datafinal = data[caracteristicas_seleccionadas_final]
datafinal
```

# Modelado

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# Realizamos la divisi√≥n para entrenamiento y prueba con los datos seleccionados
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Seleccionamos solo las caracter√≠sticas seleccionadas en x_train
x_train_seleccionado = x_train[caracteristicas_seleccionadas_final]
x_test_seleccionado = x_test[caracteristicas_seleccionadas_final]

print(x_train_seleccionado)
```

```python
# Creamos y entrenamos el modelo de regresi√≥n log√≠stica
model = LogisticRegression()
model.fit(x_train_seleccionado, y_train)
```

```python
# Hacemos predicciones
y_pred = model.predict(x_test_seleccionado)
```

# Evaluaci√≥n

```python
# Calculamos la matriz de confusi√≥n
cm = confusion_matrix(y_test, y_pred)

# Mostramos la matriz de confusi√≥n
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Presence', 'Absence'], yticklabels=['Presence', 'Absence'])
plt.xlabel('Predicci√≥n')
plt.ylabel('Verdadero')
plt.title('Matriz de Confusi√≥n')
plt.show()
```

```python
# Calculamos la precisi√≥n del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisi√≥n del modelo: {accuracy * 100:.2f}%')
```
